## What is [llustrisTNG](https://www.tng-project.org/)?



* A computer simulation of a large section of the Universe (actually, there are three different simulations, each of a differently-sized section)
* Ingredients: dark matter, regular matter, dark energy, plus gravity and other forces
* The simulation starts soon after the Big Bang, and continues until the present day
* 1 pc $=3.26$ light years


### You can watch the progress of a very similar simulation [here](https://www.illustris-project.org/movies/illustris_movie_cube_sub_frame.mp4)

## Exercises

1. Explore the images and videos on the IllustrisTNG [website](https://www.tng-project.org/media/).  Try both the Images and Videos tab and the Explore tab.

2. Watch the [video](https://www.illustris-project.org/movies/illustris_movie_cube_sub_frame.mp4) of Illustris closely.  How is the dark matter distributed? What do you notice about what happens to the dark matter distribution as time progresses?  

3. How is the gas distributed?  How does this compare to the dark matter distribution?  How does the gas distribution change as time progresses?

4. Galaxies will form in places where the matter becomes sufficiently dense. Let's say a galaxy can form once you can see a bright white color surrounded by pink in the dark matter distribution.  Approximately how many galaxies have formed in this box 2.5 billion years after the Big Bang?  How many galaxies have formed in this box 7 billion years after the Big Bang?

5. Recall each side of this box is 100 Mpc.  How many galaxies per cubic light year are there on average 2.5 billion years after the Big Bang?  And 7 billion years after the Big Bang?

6. The matter in this simulation is represented by particles.  Each particle is one million times the mass of our Sun.  How many particles would this simulation use to represent the stars in the Milky Way?  How many particles would this simulation use to represent the stars in the nearby Small Magellanic Cloud?

Now navigate to the [Explore](https://www.tng-project.org/explore/) tab, and choose the [2D](https://www.tng-project.org/explore/2d/) option.  Using the menu at the upper right, set the Base Layer to ``Stellar Light``, and the Overlay Layer to ``Dark Matter Density``.  Zoom into any object you choose so you can get a closer look.  Toggle the Overlay Layer to ``None`` and then back to ``Dark Matter Density`` so that you can see how the distributions of dark matter and stellar light differ. [Note: the stellar light corresponds to all the light that would be emitted by the stars in each simulated galaxy.]

7. Now zoom into a very low-density part of the box.  What do you notice about the similarities and differences between the dark matter and stellar light distributions?  

8. Approximately how much bigger is the dark matter component of an object in this region than its stellar light (in terms of diameter)?

9. Do all clumps of dark matter have stars in them?  If not, how would you describe the clumps that do have stars, vs. those that do not?

10. Now zoom into the highest-density part of the box.  How do your answers to the questions above change when applied to this region?  What does this tell you about the relationship between dark and stellar matter?

11. What other questions came up for you as you explored these simulations and their results?  I'm looking forward to hearing them!
